## 기초 미분적분 (calculus)

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

$y=x^3+0.5x^2-0.7x+0.1$

```python
x = np.linspace(-5,5,100) # -5부터 5까지 원소 200개를 추출해라 200차원 벡터 (200,)
def func1(x):
  return x**3 + 0.5 * x**2 - 0.7 * x + 0.1

y = func1(x)

plt.plot(x, y)
plt.grid()
```

${{dy}\over{dx}} = 3x^2+x-0.7 $

```python
# 미분 - 3차방정식을 2차로 2차방정식을 1차로...
# 미분(gradient)을 자동으로 계산해줌 (수치적 방식, 기계가 푸는 방식, numerical)
dydx = np.gradient(y, x)

plt.plot(x, dydx)
plt.grid()
```

```python
# analytic 미분 (해석적 방식, 인간이 푸는 방식)
dydx_analytic = 3 * x**2 + x - 0.7
plt.plot(x, dydx, 'r*-', label = 'Numerical')
plt.plot(x, dydx_analytic, 'b-', label = 'Analytic')
plt.grid()
plt.legend()
```

- 다중 변수함수의 미분

$z = f(x_1,x_2)$

$x_1$과 $x_2$로 $z$를 각각 미분하면?
${\partial{z}}\over{\partial{x_1}}$, ${\partial{z}}\over{\partial{x_2}}$

${{\partial{z}}\over{\partial{x_1}}} = \lim\limits_{h\to{0}}{{f(x_1+h,x_2)-f(x_1,x_2)}\over{h}}$

${{\partial{z}}\over{\partial{x_2}}} = \lim\limits_{h\to{0}}{{f(x_1,x_2+h)-f(x_1,x_2)}\over{h}}$

$f(x_1,x_2) = x_1x_2+3x_1^2-0.5x_2$

$x_1$로 미분

$x_2$로 미분

```python
x1 = np.linspace(-3, 3)
x2 = np.linspace(-3, 3)

def func(x1, x2):
  return x1 * x2 + 3 * x1**2 - 0.5 * x2

# 미분함수 정의 -> x1으로 미분
def grad_x1(x1, x2):
  # h : 1e - 10 == 10**-10
  return (func(x1+1e-10, x2) - func(x1, x2))/1e-10 
# 미분함수 정의 -> x2로 미분
def grad_x2(x1, x2):
  # h : 1e - 10 == 10**-10
  return (func(x1, x2+1e-10) - func(x1, x2))/1e-10
```

```python
# x1으로 미분한 함수와 x2로 미분한 함수 비교
grad1 = grad_x1(x1, 3)
plt.plot(x1, grad1)
plt.grid()
```

```python
# x1으로 미분한 함수와 x2로 미분한 함수 비교
grad2 = grad_x2(0.5, x2)
plt.plot(x2, grad2)
plt.grid()
```

- 미션

$ f(x_1,x_2) =x_1^2 x_2+2.5x_1 x_2^2 -1.7x_1 + 2.3x_2$

$x_1 = 2$

${{\partial{f(x_1,x_2)}}\over{\partial{x_2}}}$ 의 그래프를 -3~3까지 그려보세요

! 힌트, 미분할때 x1은 상수처럼 생각하면 되겠죠?!

```python
x1 = 2
x2 = np.linspace(-3, 3)

def func(x1, x2):
  return x1**2 * x2 + 2.5 * x1 * x2**2 - 1.7 * x1 + 2.3 * x2

def grad_x2(x1, x2):
  # h : 1e - 10 == 10**-10
  return (func(x1, x2+1e-10) - func(x1, x2))/1e-10
```

```python
grad2 = grad_x2(2, x2)
plt.plot(x2, grad2)
plt.grid()
```

    - ${dy\over{dw}}=x$
    
2. 행렬 곱미분 $Y=XW$

    - ${{\partial{Y}}\over{\partial{W}}} = {X}$
