Example.2 이차방정식 근의 공식

$ax^2+bx+c=0$ 을 만족하는 $x$는,

$x = {{-b\pm{\sqrt {b^2-4ac}}}\over{2a}}$

image.png

이차방정식의 근 정의

입력변수: $a,b,c$

출력되는 값: $x(= {{-b\pm{\sqrt {b^2-4ac}}}\over{2a}})$

- Case 1: 근이 두개인 경우 ${b^2-4ac} > 0$

- Case 2: 근이 하나인 경우  ${b^2-4ac} = 0$

- Case 3: 근이 없는 경우 ${b^2-4ac} < 0$

```
def secorder_sol(a,b,c):
  check_point = b**2-4*a*c
  if check_point > 0:
    # x**0.5 >>> 루트x
    return (-b+check_point**0.5)/(2*a), (-b-check_point**0.5)/(2*a)
  elif check_point == 0:
    return -b/(2*a)
  else:
    print("근이 없는 경우입니다.")
    return None
```

```
print(secorder_sol(1,2,0)) # 근이 두개인 경우 b^2 - 4ac > 0
print(secorder_sol(1,2,1)) # 근이 하나인 경우 b^2 - 4ac = 0
print(secorder_sol(1,2,2)) # 근이 없어요!     b^2 - 4ac < 0
print(secorder_sol(0,1,2)) # 에러가 발생하겠죠! <<<<< 2차 방정식이 아님
```
